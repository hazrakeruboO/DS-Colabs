DATA CLEANING





```python
#Converting the states in GrandElectors to Uppercase
%sql UPDATE GrandElectors SET State=UPPER(State)
%sql SELECT *FROM GrandElectors LIMIT 5;

```

```python
#Merging the two tables on the State key. We create a new table called Mergedtable1
%%sql
CREATE TABLE  Mergedtable1 AS SELECT State,Population,GrandElectors
FROM (SELECT*FROM Population INNER JOIN GrandElectors
                  ON Population.State=GrandElectors.State)
```

```python
#Displaying the Mergedtable
%%sql
SELECT *FROM Mergedtable1;
```

```python
#Now we change "District of Columbia" to "DC"
%%sql
UPDATE Mergedtable1
SET State="DC"
WHERE State="DISTRICT OF COLUMBIA";
SELECT *FROM Mergedtable1;
```

RATIO CALCULATION

```python
#We compute the ratio between the GrandElectors and the Population creating a new column to store the output
#we will first create a new column in our new table Mergedtable
%%sql
ALTER TABLE Mergedtable1 ADD Ratio;
SELECT *FROM Mergedtable1;
```

```python
#we now go ahead and compute the ratio and order the  states by decreasing ratio 
%%sql
UPDATE Mergedtable1 
SET 'Ratio'= (GrandElectors*1.0/Population*1.0);
SELECT * FROM Mergedtable1
ORDER BY Ratio DESC;
```

```python
#computing the running sum
%%sql
SELECT 
  State,GrandElectors,Population,Ratio,
  (SELECT SUM(GrandElectors) FROM Mergedtable1 AS New)
FROM Mergedtable1 AS New
```
