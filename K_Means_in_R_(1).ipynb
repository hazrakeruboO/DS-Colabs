```r
# call the libraries we will use in the lab assignment
library(tidyverse)
library(caret)
library(dplyr)
library(testthat)
library(ggplot2)
library(recipes)
library(class)

```

```r
# lets use the famous iris dataset for this example 
set.seed(123) # since there is an element of randomness to the initial centres 
data("iris") # lets have a look at the data 
str(iris) 
head(iris) 

# we will examine the first 4 dimensions to see if we can detect a different structure across the three species 
plot(iris[,1:4], col = iris[,5]) # definitely a structure there, but also some overlap 


# lets check out the kmeans function 
?kmeans 
# we can create an elbow plot in order to try and detect the number of clusters 
SST <- list() # total sum of squared error 
# run k-means for k = 1 to 10 
for (i in 1:10) SST[i] <- sum(kmeans(iris[,1:4], centers=i)$withinss) 

# plot the results 
plot(1:10, SST, pch = "", lty = 2, xaxt = "n", xlab = "Number of Centres (K)", ylab = "Total Sum of Squared Error (SST)", main = "Elbow Plot") 
abline(v=1:10, col="gray", lty=3) 
lines(x = 1:10, y = SST, lwd = 3) 
axis(side = 1, at =1:10) 

# there seems to be an "elbow" at k=3 after which there is no significant gain in increasing k 
points(x = 3, y = SST[3], pch = 1, cex = 3, col = "red", lwd = 3) 
text(x = 3+0.5, y = SST[[3]]+50, label = "Elbow!", col = "red") 

# okay so lets set k = 3 
kmeans.mod <- kmeans(iris[,1:4], centers = 3) 
# print out the summary stats 
print(kmeans.mod) 

# plots indicate kmeans has done a pretty good job 
plot(iris[,1:4], col = kmeans.mod$cluster) 

# we can examine the cluster stability by applying kmeans to a few bootstrap samples of the data 
# we then plot the centres of the results and check if they are consistent 
plot(iris$Sepal.Width, iris$Petal.Width, col = kmeans.mod$cluster, type = "n", main = "Cluster Stability", xlab = "Sepal Width", ylab = "Petal Width")
points(kmeans.mod$centers[,c(2,4)], pch = 1, cex = 3, col = "red") 
# the actual centers in red circles 
for(i in 1:10){ 
  sample <- iris[sample(nrow(iris), replace = TRUE),] 
  sample.kmeans.mod <- kmeans(sample[,1:4], centers = 3) 
  points(sample.kmeans.mod$centers[,c(2,4)], pch = 16) } 
# although subjective, these centres seem relatively stable 

# lets compare the results to the true species (even though its an unsupervised problem) 
# first relabel the three species 
levels(iris$Species)[levels(iris$Species)=="setosa"] <- median(kmeans.mod$cluster[1:50]) 
levels(iris$Species)[levels(iris$Species)=="versicolor"] <- median(kmeans.mod$cluster[51:100]) 
levels(iris$Species)[levels(iris$Species)=="virginica"] <- median(kmeans.mod$cluster[101:150]) 
# now compare 
paste("K-means Agreement = ", sum(iris$Species == kmeans.mod$cluster),"/",nrow(iris)) 
# pretty similar! ~90% 

# okay now lets re-load the data and try k-medoids 
data(iris) 
library(cluster) 
?pam 
# pam is a popular implementation of k-medoids 
kmedoids.mod <- pam(x = iris[,1:4], k = 3) 
# notice that the centres are now actual observations 
print(kmedoids.mod) 
# seems to have done a similar job 
plot(iris[,1:4], col = kmedoids.mod$clustering) 
# we can even represent the 4d data in 2d and visualise the clusters using PCA 
clusplot(kmedoids.mod, main ="2 Dimensional Representation of K-medoids Output") 
# again compare the clustering results to the actual species 
levels(iris$Species)[levels(iris$Species)=="setosa"] <- median(kmedoids.mod$clustering[1:50]) 
levels(iris$Species)[levels(iris$Species)=="versicolor"] <- median(kmedoids.mod$clustering[51:100]) 
levels(iris$Species)[levels(iris$Species)=="virginica"] <- median(kmedoids.mod$clustering[101:150]) 
paste("K-medoids Agreement = ", sum(iris$Species == kmedoids.mod$clustering),"/",nrow(iris))
# same agreement that we saw before! 

# finally lets check out an implementation of C-means 
library(e1071) 
?cmeans 
cmeans.mod <- cmeans(iris[,1:4], centers = 3) 
# cmeans output isn't quite as easy to visualise as we are given probability of membership to each cluster 
# but this output has the benefit of highlighting our uncertainty in some observations membership 
print(cmeans.mod) 

# comparing the centres of the three approaches they seem to roughly agree (in this plot at least!) 
plot(iris[,3:4], type = "n", main = "Comaparison of Cluster Centres", xlab = "Petal Length", ylab = "Petal Width") 
points(rbind(kmeans.mod$centers,kmedoids.mod$medoids,cmeans.mod$centers)[,3:4], col = c(1,1,1,2,2,2,3,3,3), pch = 19)
legend(x = "topleft", legend = c("K-means", "K-medoids", "C-means"), col = 1:3, pch = 19)
```
